```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
wine.keys()
```

```python
print(wine['DESCR'])
```

```python
data = wine['data']
feature_names = wine['feature_names']
target = wine['target']
target_names = wine['target_names']
target_names
```

```python
X = pd.DataFrame(data, columns = feature_names)
X.head()
```

```python
X.info()
```

```python
X['target'] = target
X.head()
```

```python
X_corr = X.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize = (20, 15))
sns.set(font_scale = 1.4)
sns.heatmap(X_corr, annot = True, linewidths = 0.5, cmap = 'GnBu')
plt.show()
```

```python
X_corr
```

```python
high_corr = [feature for feature, weight in zip(feature_names, X_corr['target']) if np.abs(weight) > 0.5]
high_corr
```

```python
X = X.drop(['target'], axis = 1)
X
```

```python
for feature in high_corr:
    X[feature + '_2'] = X[feature] ** 2
X.describe()
```
